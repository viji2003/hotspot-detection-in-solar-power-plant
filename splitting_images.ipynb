```python
import os
from PIL import Image


input_folder = 'internship-510/thermal _img'
output_folder = 'internship-510/split_img1'

os.makedirs(output_folder, exist_ok=True)


supported_exts = (".jpg", ".jpeg", ".png", ".bmp")

for filename in os.listdir(input_folder):
    if filename.lower().endswith(supported_exts):
        image_path = os.path.join(input_folder, filename)
        image = Image.open(image_path)

        width, height = image.size
        mid_x, mid_y = width // 2, height // 2

        parts = {
            'top_left': image.crop((0, 0, mid_x, mid_y)),
            'top_right': image.crop((mid_x, 0, width, mid_y)),
            'bottom_left': image.crop((0, mid_y, mid_x, height)),
            'bottom_right': image.crop((mid_x, mid_y, width, height))
        }

        base_name = os.path.splitext(filename)[0]


        for part_name, part_img in parts.items():
            output_filename = f"{base_name}_{part_name}.jpg"
            output_path = os.path.join(output_folder, output_filename)
            part_img.save(output_path)

print(" Splitting complete. Images saved in:", output_folder)
```
